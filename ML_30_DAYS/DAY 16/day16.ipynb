```python
import torch
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt
```

```python
X = torch.tensor([[0.,0.],[0.,1.],[1.,0.],[1.,1.]])
y = torch.tensor([[0.],[1.],[1.],[0.]])
```

```python
model = nn.Sequential(
    nn.Linear(2,4),
    nn.ReLU(),
    nn.Linear(4,1),
    nn.Sigmoid()
)

loss_fn = nn.BCELoss()
optimizer = optim.Adam(model.parameters(), lr=0.01)
```

```python
losses=[]
epochs=2000
for epoch in range(epochs):
    optimizer.zero_grad()
    outputs=model(X)
    loss=loss_fn(outputs,y)
    loss.backward()
    optimizer.step()
    losses.append(loss.item())
    if (epoch+1)%200==0:
        print(f"Epoch {epoch+1}: Loss={loss.item():.4f}")
```

```python
with torch.no_grad():
    preds=model(X)
    print("Predictions:")
    print(preds)
    print("Rounded:")
    print((preds>0.5).float())
```

```python
plt.plot(losses)
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Training Loss")
plt.show()
```
