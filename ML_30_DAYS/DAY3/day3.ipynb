```python
numeric_df = df.select_dtypes(include=['int64', 'float64'])

numeric_df.head()
```

```python
corr_matrix = numeric_df.corr()

corr_matrix
```

```python
plt.figure(figsize=(14,10))

sns.heatmap(
    corr_matrix,
    annot=True,
    cmap="coolwarm",
    fmt=".2f"
)

plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
corr_pairs = corr_matrix.unstack()

corr_pairs
```

```python
corr_pairs = corr_pairs[corr_pairs != 1]
```

```python
top_correlations = corr_pairs.sort_values(
    ascending=False
)

top_correlations.head(6)
```

```python
from scipy import stats
```

```python
home_score_z = stats.zscore(df["num__home_score"])
```

```python
df["home_score_zscore"] = home_score_z

df.head()
```

```python
outliers = df[
    (df["home_score_zscore"] > 3) |
    (df["home_score_zscore"] < -3)
]

outliers
```

```python
plt.figure(figsize=(8,5))

sns.histplot(
    df["home_score_zscore"],
    kde=True
)

plt.title("Z-score Distribution of Home Goals")

plt.show()
```
