```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from scipy import stats
```

```python
df = pd.read_csv("../DAY2/cleaned_fifa_dataset.csv")

df.head()
```

```python
match_features = df[
    [
        "num__home_score",
        "num__away_score",
        "num__home_xg",
        "num__away_xg"
    ]
]


match_features.head()
```

```python
sns.pairplot(match_features)

plt.show()
```

```python
plt.figure(figsize=(8,5))


sns.boxplot(
    y=df["num__home_score"]
)


plt.title(
    "Home Team Goals Outliers"
)


plt.ylabel(
    "Goals"
)


plt.show()
```

```python
plt.figure(figsize=(8,5))


sns.boxplot(
    y=df["num__away_score"]
)


plt.title(
    "Away Team Goals Outliers"
)


plt.ylabel(
    "Goals"
)


plt.show()
```

```python
plt.figure(figsize=(8,5))


plt.scatter(
    df["num__home_xg"],
    df["num__home_score"]
)


plt.xlabel(
    "Expected Goals (xG)"
)


plt.ylabel(
    "Actual Goals"
)


plt.title(
    "Home xG vs Home Goals"
)


plt.show()
```

```python
corr_matrix = match_features.corr()

corr_matrix
```

```python
plt.figure(figsize=(8,6))


sns.heatmap(
    corr_matrix,
    annot=True,
    fmt=".2f"
)


plt.title(
    "Football Feature Correlation Heatmap"
)


plt.show()
```

```python
Q1 = df["num__home_score"].quantile(0.25)

Q3 = df["num__home_score"].quantile(0.75)


IQR = Q3-Q1


lower_limit = Q1 - 1.5*IQR

upper_limit = Q3 + 1.5*IQR
```

```python
df["home_score_outlier"] = np.where(
    (df["num__home_score"] < lower_limit) |
    (df["num__home_score"] > upper_limit),
    1,
    0
)
```

```python
df["home_score_outlier"].value_counts()
```

```python
outlier_matches = df[
    df["home_score_outlier"] == 1
]


outlier_matches[
[
"num__home_score",
"num__away_score",
"num__home_xg",
"num__away_xg"
]
]
```

```python
plt.figure(figsize=(8,5))


plt.scatter(
    df["num__away_xg"],
    df["num__away_score"]
)


plt.xlabel(
    "Away Expected Goals"
)


plt.ylabel(
    "Away Goals"
)


plt.title(
    "Away xG vs Away Goals"
)


plt.show()
```
