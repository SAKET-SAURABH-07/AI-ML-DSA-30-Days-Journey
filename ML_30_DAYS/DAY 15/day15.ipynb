```python
relu = nn.ReLU()

x = torch.tensor([[-2.0, -1.0, 0.0, 2.0, 5.0]])

print("Before ReLU")
print(x)

print("\nAfter ReLU")
print(relu(x))
```

```python
class NeuralNetwork(nn.Module):

    def __init__(self):
        super().__init__()

        self.fc1 = nn.Linear(4, 8)
        self.relu = nn.ReLU()
        self.fc2 = nn.Linear(8, 2)

    def forward(self, x):

        x = self.fc1(x)
        x = self.relu(x)
        x = self.fc2(x)

        return x
```

```python
model = NeuralNetwork()

print(model)
```

```python
sample = torch.randn(1, 4)

print(sample)
```

```python
prediction = model(sample)

print("Prediction:")
print(prediction)
```

```python
batch = torch.randn(5, 4)

print(batch)
```

```python
predictions = model(batch)

print(predictions)
```
