```python
# ============================================================
# DAY 7 - FEATURE ENGINEERING & DIMENSIONALITY REDUCTION
# ============================================================

# ==========================
# Import Libraries
# ==========================

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.preprocessing import PolynomialFeatures
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# ==========================
# Load Dataset
# ==========================

data = load_breast_cancer()

X = pd.DataFrame(data.data, columns=data.feature_names)
y = data.target

print("Dataset Shape:", X.shape)
print()

print("First 5 Rows")
display(X.head())

# ==========================
# Dataset Information
# ==========================

print("\nDataset Info\n")
X.info()

print("\nStatistical Summary\n")
display(X.describe())

print("\nMissing Values\n")
display(X.isnull().sum())

# ============================================================
# FEATURE SELECTION - SelectKBest
# ============================================================

print("\n==============================")
print("FEATURE SELECTION - SelectKBest")
print("==============================")

selector = SelectKBest(score_func=f_classif, k=10)

X_selected = selector.fit_transform(X, y)

selected_features = X.columns[selector.get_support()]

print("\nTop 10 Selected Features:\n")

for feature in selected_features:
    print(feature)

print("\nShape Before:", X.shape)
print("Shape After :", X_selected.shape)

# ============================================================
# FEATURE ENGINEERING
# Polynomial Features
# ============================================================

print("\n==============================")
print("FEATURE ENGINEERING")
print("==============================")

poly = PolynomialFeatures(
    degree=2,
    include_bias=False
)

X_poly = poly.fit_transform(X.iloc[:, :5])

print("\nOriginal Shape :", X.iloc[:, :5].shape)
print("Polynomial Shape :", X_poly.shape)

print("\nFirst 10 Polynomial Feature Names:\n")

print(poly.get_feature_names_out(X.columns[:5])[:10])

# ============================================================
# TRAIN MODEL USING SELECTED FEATURES
# ============================================================

print("\n==============================")
print("MODEL USING SELECTED FEATURES")
print("==============================")

X_train, X_test, y_train, y_test = train_test_split(
    X_selected,
    y,
    test_size=0.2,
    random_state=42
)

model = LogisticRegression(max_iter=10000)

model.fit(X_train, y_train)

y_pred = model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)

print("\nAccuracy:", round(accuracy,4))

# ============================================================
# PCA
# ============================================================

print("\n==============================")
print("PCA")
print("==============================")

pca = PCA(n_components=2)

X_pca = pca.fit_transform(X)

print("\nOriginal Shape :", X.shape)

print("Reduced Shape :", X_pca.shape)

print("\nExplained Variance Ratio:")

print(pca.explained_variance_ratio_)

print("\nTotal Explained Variance:")

print(np.sum(pca.explained_variance_ratio_))

# ============================================================
# PCA Visualization
# ============================================================

plt.figure(figsize=(8,6))

scatter = plt.scatter(
    X_pca[:,0],
    X_pca[:,1],
    c=y,
    cmap='viridis'
)

plt.xlabel("Principal Component 1")

plt.ylabel("Principal Component 2")

plt.title("PCA 2D Projection")

plt.colorbar(scatter,label="Target")

plt.show()

# ============================================================
# Feature Importance Scores
# ============================================================

scores = selector.scores_

feature_scores = pd.DataFrame({
    "Feature":X.columns,
    "Score":scores
})

feature_scores = feature_scores.sort_values(
    by="Score",
    ascending=False
)

print("\nTop Feature Scores\n")

display(feature_scores.head(10))

# ============================================================
# Summary
# ============================================================

print("\n==============================")
print("SUMMARY")
print("==============================")

print("""
1. Feature Selection
   -> Removed less important features using SelectKBest.

2. Feature Engineering
   -> Created interaction features using PolynomialFeatures.

3. PCA
   -> Reduced dimensions from 30 features to 2 Principal Components.

4. Logistic Regression
   -> Trained using selected features.

5. Accuracy
   -> Evaluated model performance.
""")
```
